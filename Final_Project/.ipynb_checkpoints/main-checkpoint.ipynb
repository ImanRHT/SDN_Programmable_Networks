```python
class Encoder(tf.keras.layers.Layer):
    def __init__(self, vnf_vocab_size, embedding_dim, enc_units):
        super(Encoder, self).__init__()

        self.enc_units = enc_units
        self.embedding_dim = embedding_dim
        self.vnf_vocab_size = vnf_vocab_size

        # Embeddings layer is just a dictionary
        self.embeddings = tf.random.uniform(
            [self.vnf_vocab_size, self.embedding_dim], -1.0, 1.0)

        # Create RNN layer using LSTM cells
        self.lstm = tf.keras.layers.LSTM(
            self.enc_units, return_sequences=True, return_state=True, recurrent_initializer='glorot_uniform')

    def call(self, inputs, enc_state=None):
        input_vector = tf.nn.embedding_lookup(self.embeddings, inputs)

        enc_output, enc_state_h, enc_state_c = self.lstm(
            input_vector, initial_state=enc_state)

        enc_state = [enc_state_h, enc_state_c]
        return enc_output, enc_state


class BahdanauAttention(tf.keras.layers.Layer):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()

        self.units = units
        self.W1 = tf.keras.layers.Dense(units, use_bias=False)
        self.W2 = tf.keras.layers.Dense(units, use_bias=False)

        self.attention = tf.keras.layers.AdditiveAttention()

    def call(self, value, query, value_mask):

        W1_key = self.W1(value)
        W2_query = self.W2(query)

        query_mask = tf.ones(tf.shape(query)[:-1], tf.bool)

        context_vector = self.attention(inputs=[W2_query, value, W1_key],
                                        mask=[
            query_mask, value_mask])

        return context_vector


class Decoder(tf.keras.layers.Layer):
    def __init__(self, action_vocab_size, embedding_dim, dec_units):
        super(Decoder, self).__init__()

        self.dec_units = dec_units
        self.embedding_dim = embedding_dim
        self.action_vocab_size = action_vocab_size

        # Create RNN layer using LSTM cells
        self.lstm = tf.keras.layers.LSTM(
            self.dec_units, return_sequences=True, return_state=True, recurrent_initializer='glorot_uniform')

        # Add attention layer
        self.attention = BahdanauAttention(self.dec_units)

        # Obtaining attention vector from context vector
        self.Wc = tf.keras.layers.Dense(
            self.dec_units, activation=tf.math.tanh, use_bias=False)

        # Obtaining logits for each possible action
        self.fc = tf.keras.layers.Dense(self.action_vocab_size)

    def call(self, dec_input, enc_outputs, input_mask, dec_state=None):
        # Feed calculated vector to our RNN layer
        dec_output, dec_state_h, dec_state_c = self.lstm(
            dec_input, initial_state=dec_state)  # TODO do we need to iterate? if not combine with enc_output
        dec_state = [dec_state_h, dec_state_c]

        # Run attention network to obtain context vector
        context_vector = self.attention(
            enc_outputs, dec_output, input_mask)

        # Obtain attention vector [c_t, h_t]
        context_and_rnn_output = tf.concat(
            [context_vector, dec_output], axis=-1)

        # Calculate attention vector `at = tanh(Wc@[ct; ht])`
        attention_vector = self.Wc(context_and_rnn_output)

        # Calculate logits for each possible outputs (i.e. actions)
        logits = self.fc(attention_vector)

        return logits, dec_state


class ValueFunction(tf.keras.layers.Layer):
    def __init__(self, vnf_vocab_size, embedding_dim, units):
        super(ValueFunction, self).__init__()

        self.units = units
        self.embedding_dim = embedding_dim
        self.vnf_vocab_size = vnf_vocab_size

        # Embeddings layer is just a dictionary
        self.embeddings = tf.random.uniform(
            [self.vnf_vocab_size, self.embedding_dim], -1.0, 1.0)

        # Create RNN layer using LSTM cells
        self.lstm = tf.keras.layers.LSTM(
            self.units, return_sequences=True, return_state=True, recurrent_initializer='glorot_uniform')

        # MLP output layer
        self.output_layer = tf.keras.layers.Dense(1)

    def call(self, inputs):
        input_vector = tf.nn.embedding_lookup(self.embeddings, inputs)

        _, rnn_state_h, _ = self.lstm(
            input_vector)

        output = self.output_layer(rnn_state_h)

        value_estimate = tf.squeeze(output)

        return value_estimate


class Critic(tf.keras.Model):
    def __init__(self, vnf_vocab_size, embedding_dim, units):
        super(Critic, self).__init__()
        # Build layers
        self.value_function = ValueFunction(
            vnf_vocab_size, embedding_dim, units)

    def set_params(self, y_real):
        self.y_real = y_real

    def train_step(self, data):

        with tf.GradientTape() as tape:
            estimated_value = self.value_function(data)
            loss = self.loss(self.y_real, estimated_value)  # TODO placements

        # Apply an optimization step
        variables = self.trainable_variables
        gradients = tape.gradient(loss, variables)
        self.optimizer.apply_gradients(zip(gradients, variables))

        # Return a dict mapping metric names to current value
        return {'batch_loss': loss}

    def call(self, inputs):
        return self.value_function(inputs)


class Actor(tf.keras.Model):
    def __init__(self, vnf_vocab_size, action_vocab_size, embedding_dim, max_length, units):
        super(Actor, self).__init__()

        self.max_length = max_length

        # Build the encoder and decoder
        encoder = Encoder(vnf_vocab_size, embedding_dim, units)
        decoder = Decoder(action_vocab_size, embedding_dim, units)

        self.encoder = encoder
        self.decoder = decoder

    def set_params(self, lagrangian_holder, baseline_holder, placement_holder):
        self.lagrangian_holder = lagrangian_holder
        self.baseline_holder = baseline_holder
        self.placement_holder = tf.squeeze(placement_holder)

    def train_step(self, data):
        input_mask = data != 0
        outputs = []

        with tf.GradientTape() as tape:
            # Encode the input
            enc_output, enc_state = self.encoder(data)

            # Run decoder
            dec_state = enc_state

            decoder_input = tf.unstack(enc_output, num=None, axis=1)
            # TODO vocab size or action size? important check other parts
            for i in range(self.max_length):
                new_input = decoder_input[i]
                new_input = tf.expand_dims(new_input, axis=1)
                dec_output, dec_state = self.decoder(
                    new_input, enc_output, input_mask, dec_state)

                outputs.append(dec_output)

            dec_outputs_logits = tf.stack(outputs, axis=1)
            dec_outputs_logits = tf.squeeze(dec_outputs_logits)

            probs = tf.compat.v1.distributions.Categorical(
                logits=dec_outputs_logits)

            log_softmax = probs.log_prob(self.placement_holder)

            # TODO do we need this? we should enumerate
            log_softmax_mean = tf.reduce_mean(log_softmax, 1)  # [Batch]

            # Compute Loss
            advantage = self.lagrangian_holder - self.baseline_holder
            loss_rl = tf.reduce_mean(advantage * log_softmax_mean, 0)

        # Apply an optimization step
        variables = self.trainable_variables
        gradients = tape.gradient(loss_rl, variables)
        # TODO do we need clipping?
        # # Clipping
        # capped_gradients = [(tf.clip_by_norm(grad, 1.), var)
        #                     for grad, var in gradients if grad is not None]  # L2 clip
        self.optimizer.apply_gradients(zip(gradients, variables))
        # Return a dict mapping metric names to current value
        return {'batch_loss': loss_rl}

    def call(self, inputs):
        input_mask = inputs != 0  # TODO check if they are true/false
        outputs = []
        # Encode the input
        enc_output, enc_state = self.encoder(inputs)
        # Run decoder
        dec_state = enc_state

        decoder_input = tf.unstack(enc_output, num=None, axis=1)

        # TODO vocab size or action size? important check other parts (service length)
        for i in range(self.max_length):
            new_input = decoder_input[i]
            new_input = tf.expand_dims(new_input, axis=1)
            dec_output, dec_state = self.decoder(
                new_input, enc_output, input_mask, dec_state)

            outputs.append(dec_output)

        dec_outputs_logits = tf.stack(outputs, axis=1)

        # Multinomial distribution
        self.decoder_softmax = tf.nn.softmax(dec_outputs_logits)
        probs = tf.compat.v1.distributions.Categorical(
            logits=dec_outputs_logits)

        # Sample from distribution
        agent_exploration = probs.sample(1)
        agent_exploration = tf.cast(agent_exploration, tf.int32)

        return agent_exploration
```

```python
actor = Actor(len(VNFD_PROPERTIES_SIZE_SMALL),
              len(CPU_PROPERTIES_SMALL), 10, 12, 32)
critic = Critic(len(VNFD_PROPERTIES_SIZE_SMALL), 10, 32)
actor.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.99, epsilon=0.0000001),
)
critic.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss=tf.math.squared_difference,
)


""" Environment """
env = Environment(len(CPU_PROPERTIES_SMALL), len(
    VNFD_PROPERTIES_SIZE_SMALL) - 1)

""" Network service generator """
vocab_size = len(VNFD_PROPERTIES_SIZE_SMALL)  # vnfd + 1'ed
networkServices = ServiceBatchGenerator(128, 6, 12, vocab_size)
```

```python
reward_arr = []
penalty_arr = []
lagrangian_arr = []
estimated_value_arr = []
ve_loss_arr = []
agent_loss_arr = []
non_zero_arr = []
for episode in range(10000):
    # New batch of states
    networkServices.getNewState()
    input_data = tf.constant(networkServices.state)
    agent_exploration = tf.squeeze(actor(input_data), axis=3)
    estimated_value = critic(input_data)

    lagrangian, penalty, reward, constraint_occupancy, constraint_bandwidth, constraint_latency, indices = env.calculate_reward(
        networkServices, agent_exploration.numpy(), 1)

    critic.set_params(lagrangian)
    actor.set_params(lagrangian, estimated_value, agent_exploration)

    loss = critic.train_step(input_data)['batch_loss']
    loss_rl = actor.train_step(input_data)['batch_loss']
    
    estimated_value_arr.append(np.mean(estimated_value))
    reward_arr.append(np.mean(reward))
    penalty_arr.append(np.mean(penalty))
    lagrangian_arr.append(np.mean(lagrangian))
    ve_loss_arr.append(np.mean(loss))
    agent_loss_arr.append(np.mean(loss_rl))
    non_zero_arr.append(np.count_nonzero(penalty))

    if episode == 0 or episode % 100 == 0:

        print("------------")
        print("Episode: ", episode)
        print("Minibatch loss: ", agent_loss_arr[episode])

        print("Estimated Value: ", estimated_value_arr[episode])
        print("Mean reward: ", reward_arr[episode])
        print("Mean lagrangian: ", lagrangian_arr[episode])

        print("Value Estimator loss: ", ve_loss_arr[episode])
        print("Mean penalty: ", penalty_arr[episode])
        print("Count_nonzero: ", non_zero_arr[episode])
```
